```python
# for reference: 
# https://pypi.org/project/lyricsgenius/
# https://genius.com/api-clients
# https://towardsdatascience.com/song-lyrics-genius-api-dcc2819c29

import csv 
import pandas as pd
import lyricsgenius
```

```python
# loads the songs 
songs = pd.read_csv('Full_Songs_Data.csv')
# loads the token to access the Genius API 
genius = lyricsgenius.Genius('xv50TPLeT8qpbZbWIi1m35FJPb_szNvb9RRXhqn8Vy2YhexWBEhS2ndLXtN-hxu8')
# opening the file
file = open('All_popsongs_structure.csv', 'w')
writer = csv.writer(file)
#writing the header row
writer.writerow(['Top50', 'Title', 'Artist', 'Structure'])
```

```python
# finds the section titles of the word and appends them to the list 'structure'
def find_section_titles(song_lyrics, structure):
    pos1 = song_lyrics.find("[")
    pos2 = song_lyrics.find("]")
    word = song_lyrics[pos1+1:pos2]
    # cleaning up it if specifies who's singing it 
    if word.count(":") != 0:
        pos3 = word.find(":")
        word = word[0:pos3]
    structure.append(word)
```

```python
# returns the list with all of the section titles in it as strings
def return_structure(song_lyrics):
    structure = []
    loop_times = song_lyrics.count("[")
    new_lyrics = song_lyrics
    for times in range(loop_times):
        find_section_titles(new_lyrics, structure)
        pos = new_lyrics.find("]")
        new_lyrics = new_lyrics[pos + 1:]
    return structure
        
```

```python
# removes the 'Featuring ....' from the artist's name because
# that doesn't work properly with the search_song feature in the Genius API
# also replaces x or X with & because that also messes up the search_song feature
# and then returns the new name

def remove_featuring(artist):
    new_artist = artist
    if new_artist.count("Featuring") != 0:
        pos = new_artist.find("Featuring")
        new_artist = new_artist[0: pos -1]
    if new_artist.count(" x ") != 0:
        new_artist = new_artist.replace(" x ", " & ")
    if new_artist.count(" X ") != 0:
        new_artist = new_artist.replace(" X ", " & ")
    return new_artist
```

```python
# this is to automate the weird songs that just don't work 
# with the API 
def account_for_weirdness(name):
    if name == "Death Bed":
        return "Death Bed (coffee for your head)"
    if name == "ily":
        return "ily (i love you baby)"
    if "Kane Brown" in name:
        return "Kane Brown, Swae Lee & Khalid"
    return name
```

```python
for x in range(0, len(songs)): 
    rank   = songs['top50'][x]
    title  = songs['title'][x]
    # fixing stuff that doesn't work with the API
    title  = account_for_weirdness(title)    
    artist = remove_featuring(songs['artist'][x])
    # fixing stuff that doesn't work with the API
    artist = account_for_weirdness(artist)
    print("---------------------")
    try: 
        song   = genius.search_song(title, artist)
        lyrics = song.lyrics
        struc  = return_structure(lyrics)
    except:
        struc = 'COULD NOT FIND SONG'
    writer.writerow([rank, title, artist, struc])
```
