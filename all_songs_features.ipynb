```python
import csv 
import pandas as pd
```

```python
# loads the songs 
songs = pd.read_csv('All_popsongs_structure.csv')
# opening the file
file = open('All_popsong_feature_data.csv', 'w')
writer = csv.writer(file)
writer.writerow(['rank', 'num sections', 'num unique sections', 'intro or chorus beg', 'choruses', 'verses'])
```

```python
# going for 339
new_songs = songs.drop_duplicates(subset=['Title', 'Structure'], keep='last')
# for x in range(0, len(songs)): 
#     if songs['Structure'][x].count(',') == 0:
#         new_songs = songs.drop(x, axis = 0)
new_songs
```

```python
sections = ['Intro', 'Verse', 'Refrain', 'Pre-Chorus', 'Chorus', 
            'Post-Chorus', 'Interlude', 'Instrumental Bridge', 
            'Drop', 'Bridge',  'Outro']

for x in range(0, len(new_songs)): 
    rank         = songs['Top50'][x]
    num_sec      = songs['Structure'][x].count(',') + 1
    unique_sect  = 0
    intro_chorus = 0
    num_chorus   = songs['Structure'][x].count('Chorus')
    num_verse    = songs['Structure'][x].count('Verse')

    # little for-loop for getting unique number of sections
    for sect in range(len(sections)):     
        if sections[sect] in songs['Structure'][x]:
            unique_sect += 1
            
    if songs['Structure'][x].find('Verse') == 2 or songs['Structure'][x].find('Chorus') == 2:
        intro_chorus = 1
        
    if num_sec != 1:
        writer.writerow([rank, num_sec, unique_sect, intro_chorus, num_chorus, num_verse])
```

```python
file.close()
```
