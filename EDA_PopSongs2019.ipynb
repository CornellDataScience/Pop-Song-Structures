```python
import pandas as pd
import numpy as np
from collections import Counter
import statistics
from statistics import mode
```

```python
songs = pd.read_csv("/Users/akumaar/2019BillboardStructure.csv")
print(songs)
```

```python
#Add sunflower structure

sunflower_struc = "['Intro', 'Verse 1', 'Chorus', 'Verse 2', 'Chorus']"
Taki_struc = "['Intro', 'Chorus', 'Verse 1', 'Pre-Chorus', 'Chorus', 'Verse 2', 'Pre-Chorus', 'Chorus', 'Verse 3', 'Chorus']"
songs.loc[songs.index[13],'Structure'] = sunflower_struc
songs.loc[songs.index[48],'Structure'] = Taki_struc

print(songs)
#songs.loc[[13], 'Structure'] = pd.Series([sunflower_struc])
#print(songs)
```

```python
#Number of choruses, verses
rank = []
numChorus = []
numVerse = []
for i in range(len(songs)):
    rank.append(songs['Rank'][i])
    numChorus.append(songs['Structure'][i].count("'Chorus'"))
    numVerse.append(songs['Structure'][i].count("Verse"))
    
NumCV = {'rank': rank, 'numChorus': numChorus, 'numVerse':numVerse}
NumCV_DF = pd.DataFrame(data = NumCV)
print(NumCV_DF)

counter2 = 0
counter3 = 0
counter4 = 0
for i in numChorus:
    if i < 3:
        counter2 = counter2 + 1
    if i == 3:
        counter3 = counter3 +1
    if i == 4:
        counter4 = counter4 +1
        
print(counter2)
print(counter3)
print(counter4)
#All songs except rank#36 have at least 2 verses, vast majority have exactly 2 verses. 
#11 songs have exactly 2 choruses, 5 have 4 choruses, 34 have 3 choruses. Most songs have exactly 3 choruses. 
#meaning succesfull songs (34) tend to have 4 or more choruses
```

```python
#Number of individual sections

numElements = []
rank = []

for i in range(len(songs['Structure'])):
    rank.append(songs['Rank'][i])
    numElements.append(songs['Structure'][i].count(" ") + 1)
    
NumIE = {'rank': rank, 'numElements': numElements}
NumIE_DF = pd.DataFrame(data = NumIE)
print(NumIE_DF)
print(mode(numElements))
print(numElements.count(10))
print(Counter(numElements))
#noticed that every song has at least 7 elements. Most common is 10 elements which is seen 13 times
#upper limit is 14 distinct elements. 

```

```python
#Number of unique sections

numUnique = []
str = "['Verse 1', 'Pre-Chorus', 'Chorus', 'Verse 2', 'Pre-Chorus', 'Chorus', 'Bridge', 'Chorus']"
print(str.split())
for i in songs['Structure']:
    arr = i.split()
    for j in arr:
        if "1" in j:
            arr.remove(j)
        if "2" in j:
            arr.remove(j)
        if "3" in j:
            arr.remove(j)
        if "4" in j:
            arr.remove(j)
    s = set(arr)
    numUnique.append(len(s) - 2)

print(numUnique)
print(Counter(numUnique))

```
